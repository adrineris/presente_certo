```python
#ImportaÃ§Ãµes necessÃ¡riass
import google.generativeai as genai
from google.colab import userdata
from IPython.display import display
from ipywidgets import widgets, interact, fixed
from ipywidgets.widgets import VBox, HBox
import os
import functools

# ConfiguraÃ§Ãµes da API Gemini
GOOGLE_API_KEY=userdata.get('API_KEY')
genai.configure(api_key=GOOGLE_API_KEY)
model = genai.GenerativeModel(model_name='gemini-1.0-pro')
chat = model.start_chat(history=[])

#  Gera um prompt com base nas informaÃ§Ãµes do usuÃ¡rio e interage com a API Gemini para obter sugestÃµes de presentes.
def gerar_prompt(ocasiao, idade, genero, relacao, interesses, orcamento, excluir, informacao_adicional):

  # Concatenar as informaÃ§Ãµes do usuÃ¡rio em um Ãºnico prompt
  prompt = f"""
  ## Prompt para presente

  **OcasiÃ£o:** {ocasiao}
  **Idade:** {idade} anos
  **GÃªnero:** {genero}
  **RelaÃ§Ã£o:** {relacao}
  **Interesses:** {", ".join( interesses)}
  **OrÃ§amento:** Aproximadamente R${orcamento}
  **ExclusÃµes:** {excluir}
  **InformaÃ§Ã£o adicional:** {informacao_adicional}

  ## Gere sugestÃµes de presentes para a pessoa descrita acima, considerando a ocasiÃ£o, idade, gÃªnero, interesses, orÃ§amento e informaÃ§Ãµes adicionais.
  """
  return prompt
#FunÃ§Ã£o para enviar o prompt para a API Gemini e obter as sugestÃµes de presentes
def enviar(prompt):
  resposta = chat.send_message(prompt)
  # Extrair as sugestÃµes de presentes da resposta da API
  sugestoes_de_presentes = resposta.text
  # Retornar as sugestÃµes de presentes
  return sugestoes_de_presentes

# Criando widgets para interaÃ§Ã£o com o usuÃ¡rio
ocasiao = widgets.Text()
idade = widgets.IntSlider(min=1, max=100)
genero = widgets.RadioButtons(options=["Masculino", "Feminino", "Outro"])
relacao = widgets.Text()
interesse1 = widgets.Text(description="1:")
interesse2 = widgets.Text(description="2:")
interesse3 = widgets.Text(description="3")
interesse4 = widgets.Text(description="4:")
interesse5 = widgets.Text(description="5:")

orcamento = widgets.IntSlider(min=0, max=1000)
excluir = widgets.Text()
informacao_adicional = widgets.Textarea()

interesses = VBox([interesse1, interesse2, interesse3, interesse4, interesse5])

# Apresentando os widgets ao usuÃ¡rio
print("Qual Ã© a ocasiÃ£o? (Exemplo: AniversÃ¡rio, Natal, Dia das mÃ£es): ")
display(ocasiao)
print("\nQuantos anos a pessoa tem?")
display(idade)
print("\nQual o genero da pessoa?")
display(genero)
print("\nQual sua relaÃ§Ã£o com a pessoa? (Exemplo: filho, pai, mÃ£e, esposo)?")
display(relacao)
print("\nQuais sÃ£o os interesses da pessoa? (Exemplo: pilates, moda, cÃ£es, uma religiosidade especÃ­fica, viagens, etc)")
display(interesses)
print("\nQual o orÃ§amento mÃ©dio (R$)? ")
display(orcamento)
print("\nQuais presentes nÃ£o sugerir? ")
display(excluir)
print("\nOutras informaÃ§Ãµes que vocÃª gostoria de prestar?")
display(informacao_adicional)

# Criando botÃ£o para gerar sugestÃµes e Ã¡rea de output para exibir os resultados
button = widgets.Button(description="Obter sugestÃ£o")
output = widgets.Output()

display(button, output)

def on_button_clicked(b):
    with output:
        interesses_values = [interesse.value for interesse in interesses.children]
        prompt=gerar_prompt(ocasiao.value, idade.value, genero.value, relacao.value, interesses_values, orcamento.value, excluir.value, informacao_adicional.value)
        print(prompt)
        resposta=enviar(prompt)
        print(resposta)
button.on_click(on_button_clicked)
```
