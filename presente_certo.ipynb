```python
#Importações necessáriass
import google.generativeai as genai
from google.colab import userdata
from IPython.display import display
from ipywidgets import widgets, interact, fixed
from ipywidgets.widgets import VBox, HBox
import os
import functools

# Configurações da API Gemini
GOOGLE_API_KEY=userdata.get('API_KEY')
genai.configure(api_key=GOOGLE_API_KEY)
model = genai.GenerativeModel(model_name='gemini-1.0-pro')
chat = model.start_chat(history=[])

#  Gera um prompt com base nas informações do usuário e interage com a API Gemini para obter sugestões de presentes.
def gerar_prompt(ocasiao, idade, genero, relacao, interesses, orcamento, excluir, informacao_adicional):

  # Concatenar as informações do usuário em um único prompt
  prompt = f"""
  ## Prompt para presente

  **Ocasião:** {ocasiao}
  **Idade:** {idade} anos
  **Gênero:** {genero}
  **Relação:** {relacao}
  **Interesses:** {", ".join( interesses)}
  **Orçamento:** Aproximadamente R${orcamento}
  **Exclusões:** {excluir}
  **Informação adicional:** {informacao_adicional}

  ## Gere sugestões de presentes para a pessoa descrita acima, considerando a ocasião, idade, gênero, interesses, orçamento e informações adicionais.
  """
  return prompt
#Função para enviar o prompt para a API Gemini e obter as sugestões de presentes
def enviar(prompt):
  resposta = chat.send_message(prompt)
  # Extrair as sugestões de presentes da resposta da API
  sugestoes_de_presentes = resposta.text
  # Retornar as sugestões de presentes
  return sugestoes_de_presentes

# Criando widgets para interação com o usuário
ocasiao = widgets.Text()
idade = widgets.IntSlider(min=1, max=100)
genero = widgets.RadioButtons(options=["Masculino", "Feminino", "Outro"])
relacao = widgets.Text()
interesse1 = widgets.Text(description="1:")
interesse2 = widgets.Text(description="2:")
interesse3 = widgets.Text(description="3")
interesse4 = widgets.Text(description="4:")
interesse5 = widgets.Text(description="5:")

orcamento = widgets.IntSlider(min=0, max=1000)
excluir = widgets.Text()
informacao_adicional = widgets.Textarea()

interesses = VBox([interesse1, interesse2, interesse3, interesse4, interesse5])

# Apresentando os widgets ao usuário
print("Qual é a ocasião? (Exemplo: Aniversário, Natal, Dia das mães): ")
display(ocasiao)
print("\nQuantos anos a pessoa tem?")
display(idade)
print("\nQual o genero da pessoa?")
display(genero)
print("\nQual sua relação com a pessoa? (Exemplo: filho, pai, mãe, esposo)?")
display(relacao)
print("\nQuais são os interesses da pessoa? (Exemplo: pilates, moda, cães, uma religiosidade específica, viagens, etc)")
display(interesses)
print("\nQual o orçamento médio (R$)? ")
display(orcamento)
print("\nQuais presentes não sugerir? ")
display(excluir)
print("\nOutras informações que você gostoria de prestar?")
display(informacao_adicional)

# Criando botão para gerar sugestões e área de output para exibir os resultados
button = widgets.Button(description="Obter sugestão")
output = widgets.Output()

display(button, output)

def on_button_clicked(b):
    with output:
        interesses_values = [interesse.value for interesse in interesses.children]
        prompt=gerar_prompt(ocasiao.value, idade.value, genero.value, relacao.value, interesses_values, orcamento.value, excluir.value, informacao_adicional.value)
        print(prompt)
        resposta=enviar(prompt)
        print(resposta)
button.on_click(on_button_clicked)
```
